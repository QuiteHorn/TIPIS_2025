```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd

adult = fetch_ucirepo(id=2) 
  
X = adult.data.features 
y = adult.data.targets 

df = pd.concat([X, y], axis=1)
```

```python
print(f"1) Число столбцов в наборе данных {len(df.columns)}")
```

```python
print(f"2) Есть ли пропуски в данных? Если есть, то в каких столбцах")
columnsWithNull = df.columns[df.isna().any()].tolist()
print(columnsWithNull)
```

```python
print(f"3) Кол-во уникальных значений в столбце race {df['race'].nunique()}")
```

```python
print(f"4) Медиана hours-per-week {df['hours-per-week'].median()}")
```

```python
print(f"5) Кого больше - женщин или мужчин с ЗП >50K?")

male = len(df[(df['income'] == '>50K') & (df['sex'] == 'Male')])
female = len(df[(df['income'] == '>50K') & (df['sex'] == 'Female')])

if male > female:
    print("Мужчин")
elif female > male:
    print("Женщин")
else:
    print("Поровну")
```

```python
print(f"6) Заполните пропущенные данные в отдельных столбцах наиболее встречаемыми значениями. Как еще можно было бы заполнить пропущенные данные?")

df = df.fillna(df.mode().iloc[0])
df
```
